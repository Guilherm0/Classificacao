```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/vendas.xlsx')
df.head()
```

```python
X = df.drop('cliente_comprou', axis=1)

y = df['cliente_comprou']
```

```python
X.head()
```

```python
X = pd.get_dummies(data=X, prefix='test')
```

```python
X.head()
```

```python
from sklearn import tree

arvore = tree.DecisionTreeClassifier()

modelo = arvore.fit(X, y)
```

```python
modelo
```

```python
fig = plt.figure(figsize=(70,30))

fig = tree.plot_tree(
    modelo,
    feature_names= X.columns,
    class_names=['yes', 'no'],
    filled = True
)
```
